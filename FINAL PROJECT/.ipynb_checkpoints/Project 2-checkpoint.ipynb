#  House Price Prediction

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
```

```python
df=pd.read_csv('HousePrice.csv')
```

```python
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.boxplot(df.medv, orient='h')
plt.subplot(1,2,2)
sns.kdeplot(df.medv)
```

```python
count=1
plt.figure(figsize=(15,10))
for i in np.arange(0,len(df.columns)-1):
    plt.subplot(3,4,count)
    sns.kdeplot(df.iloc[:,i])
    count=count+1
```

```python
df.corr()
```

```python
sns.pairplot(df)
```

```python
X=df.drop('medv',axis=1)
y=df.medv
X2=sm.add_constant(X)
lm1=sm.OLS(y,X2).fit()
print(lm1.summary())
```

```python
formula='medv~crim+zn+nox+rm+dis+rad+tax+ptratio+lstat'
lm2=smf.ols(formula=formula, data=df).fit()
print(lm2.summary())
```

```python
pred=lm2.predict(df)
```

```python
sns.scatterplot(x=df.medv,y=pred)
```

```python
sns.scatterplot(x=df.medv,y=pred-df.medv)
```

# Scaled Model

```python
def scale(data):
    dataout=(data - data.min())/(data.max() -data.min())
    return dataout
```

```python
df.columns
```

```python
df.info()
```

```python
for i in df.columns:
    df[i]=scale(df[i])
```

```python
df.describe()
```

```python
formula='medv~crim+zn+nox+rm+dis+rad+tax+ptratio+lstat'
lm3=smf.ols(formula=formula, data=df).fit()
print(lm3.summary())
```

```python
lm3.params.abs().sort_values(ascending=False)
```

```python
lm3.params.sort_values(ascending=False)
```
