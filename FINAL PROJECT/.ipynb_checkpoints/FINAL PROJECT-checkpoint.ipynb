# ANOVA

```python
sns.boxplot(x=df.Calories, y=df.Milk)
```

```python
formula='Calories~C(Milk)'
model=ols(formula,df).fit()
aov_table=anova_lm(model)
print(aov_table)
```

```python
dfnew=df[df.Milk!='None']
dfnew.shape,df.shape
```

```python
sns.boxplot(x=dfnew.Calories, y= dfnew.Milk)
```

```python
formula='Calories~C(Milk)'
model=ols(formula,dfnew).fit()
aov_table=anova_lm(model)
print(aov_table)
```

```python
res=stat()
res.tukey_hsd(dfnew,
             res_var='Calories',
             xfac_var='Milk',
             anova_model='Calories~C(Milk)')
res.tukey_summary
```

```python
sns.boxplot(x=dfnew.Calories, y= dfnew.Milk)
```

# Linear Regression 

```python
df.columns
```

```python
formula='Calories~Category+Beverage+Size+Milk+TotalFat+TransFat+SaturatedFat+Sodium+TotalCarb+Cholesterol+DietaryFibre+Sugars+Protein'
lm=smf.ols(formula,data=df).fit()
print(lm.summary())
```

```python
formula='Calories~TotalFat+Sugars+Protein'
lm=smf.ols(formula=formula, data=df).fit()
print(lm.summary())
```

```python
formula='Calories~TransFat+Sugars+Protein'
lm=smf.ols(formula=formula, data=df).fit()
print(lm.summary())
```

```python
pred=lm.predict(df)
```

```python
sns.scatterplot(x=df.Calories,y=pred)
```

```python
sns.scatterplot(x=df.Calories,y=pred-df.Calories)
```

```python
sns.kdeplot(pred-df.Calories)
```
