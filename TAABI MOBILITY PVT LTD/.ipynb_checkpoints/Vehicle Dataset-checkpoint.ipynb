```python
df.isnull().sum()
```

```python
# To handle missing values, we will drop the columns with all the null values in it.
```

```python
newdf=df.dropna(axis='columns',how='all')
newdf
```

```python
newdf.isnull().sum().head(33)
```

```python
#The runtime has 14 null values, to handle it, imputing the runtime column with mean of runtime.
#The fuel level has 2 null values, to handle it, imputing the null values in the column with mean of fuel level.
#The adblue level has 3664 null values, to handle it, imputing the null values in the column with mean of adblue level.
```

```python
newdf.adblue_level.describe()
```

```python
newdf.fuel_level.describe()
```

```python
newdf.runtime.describe()
```

```python
#Now we can impute the data with mean values in the columns with null values.
```

```python
newdf['runtime'].fillna(int(newdf['runtime'].mean()), inplace=True)
newdf
```

```python
newdf['fuel_level'].fillna(int(newdf['fuel_level'].mean()), inplace=True)
```

```python
newdf['adblue_level'].fillna(int(newdf['adblue_level'].mean()), inplace=True)
newdf
```
